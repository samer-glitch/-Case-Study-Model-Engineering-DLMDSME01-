```python
import pandas as pd

# Load the dataset
transactions = pd.read_csv('PSP_Jan_Feb_2019.csv')

# Convert 'tmsp' to datetime format
transactions['tmsp'] = pd.to_datetime(transactions['tmsp'])

# Extracting time-related features
transactions['second'] = transactions['tmsp'].dt.second
transactions['minute'] = transactions['tmsp'].dt.minute
transactions['hour'] = transactions['tmsp'].dt.hour
transactions['day_of_week'] = transactions['tmsp'].dt.dayofweek

# Adding feature for number of tries to get a successful transaction
# Assuming each row in the dataset represents a transaction attempt
transactions['number_of_tries'] = transactions.groupby(['PSP', 'country', 'amount', 'card']).cumcount() + 1
# Saving the modified dataset
# Saving the modified dataset in the current working directory
transactions.to_csv('PSP_Jan_Feb_2019modfied1_with_tries.csv', index=False)
# Display the first few rows of the modified dataset
transactions.head(20)
```
